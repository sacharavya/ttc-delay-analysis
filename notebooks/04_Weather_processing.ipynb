```python
import math
import pandas as pd
from pathlib import Path

DATA_DIR = Path('../data')
WEATHER_DIR = DATA_DIR / 'raw/weather'

cc = pd.read_csv(WEATHER_DIR / 'climate_combined.csv')
wc = pd.read_csv(WEATHER_DIR / 'weather_complete.csv', low_memory=False)

cc['Date/Time (LST)'] = pd.to_datetime(cc['Date/Time (LST)'])
wc['Date/Time (LST)'] = pd.to_datetime(wc['Date/Time (LST)'])

# Auto-find lat/lon columns (handles encoding differences in column names)
def find_col(df, keyword):
    matches = [c for c in df.columns if keyword.lower() in c.lower()]
    return matches[0] if matches else None

cc_lon = find_col(cc, 'longitude')
cc_lat = find_col(cc, 'latitude')
wc_lon = find_col(wc, 'longitude')
wc_lat = find_col(wc, 'latitude')

KEY_COLS = [
    'Temp (°C)', 'Dew Point Temp (°C)', 'Rel Hum (%)',
    'Wind Dir (10s deg)', 'Wind Spd (km/h)',
    'Visibility (km)', 'Stn Press (kPa)',
    'Wind Chill', 'Hmdx', 'Weather'
]

print("=" * 65)
print(f"{'ATTRIBUTE':<30} {'climate_combined':>16} {'weather_complete':>16}")
print("=" * 65)

print(f"{'Rows':<30} {len(cc):>16,} {len(wc):>16,}")
print(f"{'Columns':<30} {cc.shape[1]:>16} {wc.shape[1]:>16}")
print(f"{'Station':<30} {'City Centre':>16} {'Pearson Intl':>16}")
print(f"{'Climate ID':<30} {str(cc['Climate ID'].iloc[0]):>16} {str(wc['Climate ID'].iloc[0]):>16}")
print(f"{'Latitude':<30} {cc[cc_lat].iloc[0]:>16.2f} {wc[wc_lat].iloc[0]:>16.2f}")
print(f"{'Longitude':<30} {cc[cc_lon].iloc[0]:>16.2f} {wc[wc_lon].iloc[0]:>16.2f}")
print(f"{'Date start':<30} {str(cc['Date/Time (LST)'].min())[:16]:>16} {str(wc['Date/Time (LST)'].min())[:16]:>16}")
print(f"{'Date end':<30} {str(cc['Date/Time (LST)'].max())[:16]:>16} {str(wc['Date/Time (LST)'].max())[:16]:>16}")

cc25 = cc[cc['Year'] == 2025]
wc25 = wc[wc['Year'] == 2025]
expected = 365 * 24

print(f"{'2025 rows':<30} {len(cc25):>16,} {len(wc25):>16,}")
print(f"{'2025 months covered':<30} {len(cc25['Month'].unique()):>16} {len(wc25['Month'].unique()):>16}")
print(f"{'2025 hour coverage %':<30} {len(cc25)/expected*100:>15.1f}% {len(wc25)/expected*100:>15.1f}%")

print(f"\\n── Missing rate on key columns (2025 only) ──")
print(f"  {'Column':<28} {'climate_combined':>16} {'weather_complete':>16}")
print("  " + "-" * 62)
for col in KEY_COLS:
    cc_miss = f"{cc25[col].isnull().mean()*100:.1f}%" if col in cc25.columns else "N/A"
    wc_miss = f"{wc25[col].isnull().mean()*100:.1f}%" if col in wc25.columns else "N/A"
    print(f"  {col:<28} {cc_miss:>16} {wc_miss:>16}")

cc_only = set(cc.columns) - set(wc.columns)
wc_only = set(wc.columns) - set(cc.columns)
print(f"\\n── Columns unique to climate_combined : {cc_only}")
print(f"── Columns unique to weather_complete : {wc_only}")

lat1, lon1 = cc[cc_lat].iloc[0], cc[cc_lon].iloc[0]
lat2, lon2 = wc[wc_lat].iloc[0], wc[wc_lon].iloc[0]
dlat = math.radians(lat2 - lat1)
dlon = math.radians(lon2 - lon1)
a = (math.sin(dlat/2)**2 +
     math.cos(math.radians(lat1)) * math.cos(math.radians(lat2)) * math.sin(dlon/2)**2)
dist_km = 6371 * 2 * math.atan2(math.sqrt(a), math.sqrt(1 - a))
print(f"\\n── Distance between stations: {dist_km:.1f} km")
```

```python
# ── Cell 5: Clean, merge, feature-engineer → save weather_2025.parquet ───
import pandas as pd
import numpy as np
from pathlib import Path

DATA_DIR  = Path('../data')
WEATHER_DIR = DATA_DIR / 'raw/weather'
OUT_DIR   = DATA_DIR / 'processed'
OUT_DIR.mkdir(parents=True, exist_ok=True)

# ── Load both ─────────────────────────────────────────────────────────────
wc = pd.read_csv(WEATHER_DIR / 'weather_complete.csv',  low_memory=False)
cc = pd.read_csv(WEATHER_DIR / 'climate_combined.csv')

# Fix BOM in climate_combined longitude column
cc.columns = [c.lstrip('\\ufeff').strip('"') for c in cc.columns]

# Parse datetime
wc['datetime'] = pd.to_datetime(wc['Date/Time (LST)'])
cc['datetime'] = pd.to_datetime(cc['Date/Time (LST)'])

# ── Filter both to 2025 ───────────────────────────────────────────────────
wc25 = wc[wc['Year'] == 2025].copy().reset_index(drop=True)
cc25 = cc[cc['Year'] == 2025].copy().reset_index(drop=True)

wc25 = wc25.set_index('datetime')
cc25 = cc25.set_index('datetime')

# ── Core columns we care about ────────────────────────────────────────────
CORE = ['Temp (°C)', 'Dew Point Temp (°C)', 'Rel Hum (%)',
        'Wind Dir (10s deg)', 'Wind Spd (km/h)',
        'Visibility (km)', 'Stn Press (kPa)',
        'Precip. Amount (mm)', 'Wind Chill', 'Weather']

# ── Use Pearson as base, fill gaps from City Centre ───────────────────────
weather = wc25[CORE].copy()
for col in CORE:
    if col in cc25.columns:
        mask = weather[col].isnull()
        weather.loc[mask, col] = cc25.loc[mask.index.intersection(cc25.index), col]

print(f"After merge — missing values:")
print(weather[CORE].isnull().sum())

# ── Feature engineering ───────────────────────────────────────────────────
# 1. Precipitation flags
weather['precip_mm']    = weather['Precip. Amount (mm)'].fillna(0)
weather['is_raining']   = weather['Weather'].str.contains(
                            r'[Rr]ain|[Dd]rizzle', na=False).astype(int)
weather['is_snowing']   = weather['Weather'].str.contains(
                            r'[Ss]now|[Bb]lizzard|[Ff]lurr', na=False).astype(int)
weather['is_fog']       = weather['Weather'].str.contains(
                            r'[Ff]og|[Mm]ist', na=False).astype(int)
weather['is_precip']    = ((weather['is_raining'] == 1) |
                            (weather['is_snowing'] == 1)).astype(int)

# 2. Wind speed category
weather['wind_cat'] = pd.cut(
    weather['Wind Spd (km/h)'].fillna(0),
    bins=[-1, 10, 20, 40, 9999],
    labels=['calm', 'light', 'moderate', 'strong']
)

# 3. Feels-like: use Wind Chill in winter, Humidex in summer, else Temp
weather['feels_like'] = np.where(
    weather['Wind Chill'].notna(),  weather['Wind Chill'],
    np.where(weather['Hmdx'].notna() if 'Hmdx' in weather.columns else False,
             weather.get('Hmdx', weather['Temp (°C)']),
             weather['Temp (°C)'])
)

# 4. Temperature bins
weather['temp_bin'] = pd.cut(
    weather['Temp (°C)'],
    bins=[-50, -10, 0, 5, 15, 25, 50],
    labels=['extreme_cold', 'cold', 'cool', 'mild', 'warm', 'hot']
)

# 5. Low visibility flag (< 5 km — affects driving conditions)
weather['low_visibility'] = (weather['Visibility (km)'] < 5).astype(int)

# 6. Date column for joining to delay data
weather['date'] = weather.index.date
weather['hour'] = weather.index.hour

print(f"\\nFinal shape: {weather.shape}")
print(f"\\nSample of engineered features:")
print(weather[['Temp (°C)', 'feels_like', 'temp_bin',
               'Wind Spd (km/h)', 'wind_cat',
               'is_raining', 'is_snowing', 'is_fog',
               'precip_mm', 'low_visibility']].head(8).to_string())

# ── Save ──────────────────────────────────────────────────────────────────
out_path = OUT_DIR / 'weather_2025.parquet'
weather.reset_index().to_parquet(out_path, index=False)
print(f"\\n✓ Saved → {out_path}  ({out_path.stat().st_size/1e6:.2f} MB)")
```
