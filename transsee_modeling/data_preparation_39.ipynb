```python
import pandas as pd
```

```python
# Load the processed data
print("Loading parquet data...")
df = pd.read_parquet("./tmp/unfv_with_weather_route_29_39.parquet")
print(f"Columns: {df.columns.tolist()}")
```

```python
df_39 = df[df["route"] == 39]
```

```python
df_39.dtypes
```

```python
df_39[df_39.stopid.str.contains("_")].stopid.value_counts().sum()
```

```python
def create_bound(stopid):
    if "_" in stopid:
        get_bound = stopid.split("_")[1]
        return int(get_bound)
    else:
        return 0

df_39['bound'] = df_39['stopid'].apply(create_bound)
```

```python
df_39[df_39.stopid.str.contains("_")].bound.value_counts().sum()
```

```python
df_39['stopid'] = df_39['stopid'].apply(lambda x: int(x.split("_")[0]) if "_" in x else int(x))
```

```python
df_39.isna().sum()
```

```python
df_39_no_na_dropped = df_39.dropna(subset=['Climate ID', 'thrusteet'])
```

```python
df_39_no_na_dropped.shape
```

```python
df_39_no_na_dropped.isna().sum()
```

```python
df_39_no_na_dropped.to_csv("./tmp/df_39_no_na_dropped_without_encoding.csv", index=False)
```

```python
df_39_no_na_dropped['thrustreet'] = df_39_no_na_dropped['thrusteet']
df_39_no_na_dropped_encoding = pd.get_dummies(df_39_no_na_dropped, columns=['thrusteet'], prefix=['thrusteet'], dtype=int)
```

```python
df_39_no_na_dropped_encoding.head(), df_39_no_na_dropped_encoding.shape
```

```python
df_39_no_na_dropped_encoding.to_csv("./tmp/unfv_route_39_no_na_dropped_encoding.csv", index=False)
```
